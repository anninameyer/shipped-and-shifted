```python
import pandas as pd
import numpy as np
from skbio.stats.composition import multiplicative_replacement, clr

def prepare_clr(df, delta=1e-8, tss=False):
    """Apply multiplicative replacement and CLR transform."""
    if tss:
        df = df.div(df.sum(axis=1), axis=0)
    mat = multiplicative_replacement(df.values, delta=delta)
    return pd.DataFrame(clr(mat), index=df.index, columns=df.columns)
```

```python
# Apply to your actual data
microbe_clr = prepare_clr(otu_df_rel, tss=False)     # already relative?
metab_clr   = prepare_clr(metab_df, tss=True)        # raw metabolite intensities
microbe_clr_ITS = prepare_clr(otu_df_ITS_rel, tss=True)     # already relative?
```

```python
def clr_abundance_weighted_interaction_matrix(A_clr, B_clr, threshold=0):
    """
    Computes fully abundance-weighted CLR-based co-occurrence between A and B.
    Returns: matrix (rows=A.columns, cols=B.columns)
    """
    result = pd.DataFrame(index=A_clr.columns, columns=B_clr.columns)

    for a in A_clr.columns:
        for b in B_clr.columns:
            x = A_clr[a]
            y = B_clr[b]

            # Compute abundance-weighted product
            joint_weight = x * y

            # Optional: filter small contributions
            joint_weight[abs(x) < threshold] = 0
            joint_weight[abs(y) < threshold] = 0

            weights = abs(joint_weight)
            numerator = (x * y * weights).sum()
            denominator = weights.sum()

            result.loc[a, b] = numerator / denominator if denominator > 0 else 0

    return result.astype(float)
```

```python
# Microbe–Microbe co-occurrence
microbe_matrix = clr_abundance_weighted_interaction_matrix(microbe_clr, microbe_clr, threshold=0)

# Metabolite–Metabolite co-occurrence
metab_matrix = clr_abundance_weighted_interaction_matrix(metab_clr, metab_clr, threshold=0)

# Microbe–Metabolite directional interaction
cross_matrix = clr_abundance_weighted_interaction_matrix(microbe_clr, metab_clr, threshold=0)

cross_microbe_matrix = clr_abundance_weighted_interaction_matrix(microbe_clr, microbe_clr_ITS, threshold=0)

microbe_matrix2 = clr_abundance_weighted_interaction_matrix(microbe_clr_ITS, microbe_clr_ITS, threshold=0)

cross_matrix2 = clr_abundance_weighted_interaction_matrix(microbe_clr_ITS, metab_clr, threshold=0)
```

```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.drawing.nx_agraph import graphviz_layout

import random



def jitter_positions(pos, scale=10, seed=None):
    """Add small random noise to node positions with optional reproducibility."""
    if seed is not None:
        random.seed(seed)
    return {
        node: (x + random.uniform(-scale, scale), y + random.uniform(-scale, scale))
        for node, (x, y) in pos.items()
    }



# Threshold to include an edge
THRESHOLD = 0.2

# Create an empty undirected graph
G = nx.Graph()

# --- Microbe–Microbe edges ---
for i in microbe_microbe_matrix.index:
    for j in microbe_microbe_matrix.columns:
        if i < j:  # avoid duplicates
            score = microbe_microbe_matrix.loc[i, j]
            if abs(score) >= THRESHOLD:
                G.add_edge(i, j, weight=score)

# --- Metabolite–Metabolite edges ---
for i in metabolite_metabolite_matrix.index:
    for j in metabolite_metabolite_matrix.columns:
        if i < j:
            score = metabolite_metabolite_matrix.loc[i, j]
            if abs(score) >= THRESHOLD:
                G.add_edge(i, j, weight=score)

# --- Microbe–Metabolite edges ---
for microbe in microbe_metabolite_matrix.index:
    for metabolite in microbe_metabolite_matrix.columns:
        score = microbe_metabolite_matrix.loc[microbe, metabolite]
        if abs(score) >= THRESHOLD:
            G.add_edge(microbe, metabolite, weight=score)

# --- ITS Microbe–ITS Microbe edges ---
for i in microbe_matrix2.index:
    for j in microbe_matrix2.columns:
        if i < j:
            score = microbe_matrix2.loc[i, j]
            if abs(score) >= THRESHOLD:
                G.add_edge(i, j, weight=score)

# --- ITS Microbe–Metabolite edges ---
for microbe in cross_matrix2.index:
    for metabolite in cross_matrix2.columns:
        score = cross_matrix2.loc[microbe, metabolite]
        if abs(score) >= THRESHOLD:
            G.add_edge(microbe, metabolite, weight=score)

# --- ITS Microbe–16S Microbe edges ---
for microbe1 in cross_microbe_matrix.index:
    for microbe2 in cross_microbe_matrix.columns:
        score = cross_microbe_matrix.loc[microbe1, microbe2]
        if abs(score) >= THRESHOLD:
            G.add_edge(microbe1, microbe2, weight=score)

# === Network Metrics ===
# Calculate network metrics
degree_centrality = nx.degree_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G, weight='weight')
closeness_centrality = nx.closeness_centrality(G)
eigenvector_centrality = nx.eigenvector_centrality_numpy(G, weight='weight')

# Store in DataFrame
import pandas as pd
metrics_df = pd.DataFrame({
    'DegreeCentrality': pd.Series(degree_centrality),
    'BetweennessCentrality': pd.Series(betweenness_centrality),
    'ClosenessCentrality': pd.Series(closeness_centrality),
    'EigenvectorCentrality': pd.Series(eigenvector_centrality)
})

# Sort and inspect top nodes
metrics_df_sorted = metrics_df.sort_values(by='DegreeCentrality', ascending=False)
print(metrics_df_sorted.head(10))

# Optional: save to CSV
metrics_df.to_csv("stats/20250523_network_centralities_co_correlation_4C.csv")


# --- Node attributes ---
microbes = microbe_microbe_matrix.index.tolist()
metabolites = metabolite_metabolite_matrix.index.tolist()
microbe_ITS_nodes = microbe_microbe_matrix2.index.tolist()


# Node sizes still based on abundance
node_sizes = []
for node in G.nodes:
    if node in microbes:
        node_sizes.append(microbe_abund.get(node, 1))
    elif node in metabolites:
        node_sizes.append(metabolite_abund.get(node, 1))
    else:
        node_sizes.append(1)  # fallback

# Use color dictionary
node_colors = [color_dict.get(node, 'gray') for node in G.nodes]

# Rescale sizes (keep this if desired)
node_sizes = [300 + (s / max(node_sizes)) * 1700 for s in node_sizes]


# --- Edge attributes: color, thickness, Graphviz spring length ---
for u, v, d in G.edges(data=True):
    w = d["weight"]
    d["penwidth"] = 1 + 0.4 * abs(w)
    #d["color"] = "#CB6BCEFF" if w > 0 else "#345084FF"
    d["color"] = "#345084FF" if w > 0 else "#C0C0C0"
    d["len"] = 1 / abs(w) if w > 0 else 2 + abs(w)  # Stronger positives pull close, negatives push apart
    d["label"] = f"{w:.2f}"

# --- Layout with Graphviz (neato) ---
pos = graphviz_layout(G, prog="neato")  # 'neato' handles spring-like layouts well

pos = jitter_positions(pos, scale=6, seed = 42)  # adjust scale as needed (e.g., 5–20)


# --- Draw ---
# --- Separate edges by sign ---
pos_edges = [(u, v) for u, v, d in G.edges(data=True) if d["weight"] > 0]
neg_edges = [(u, v) for u, v, d in G.edges(data=True) if d["weight"] < 0]

# Colors and widths for each group
pos_colors = [G[u][v]["color"] for u, v in pos_edges]
pos_widths = [G[u][v]["penwidth"] for u, v in pos_edges]

neg_colors = [G[u][v]["color"] for u, v in neg_edges]
neg_widths = [G[u][v]["penwidth"] for u, v in neg_edges]

# --- Draw the network ---
plt.figure(figsize=(13, 8))

# Draw edges in alpha layers
nx.draw_networkx_edges(G, pos, edgelist=neg_edges, edge_color=neg_colors, width=neg_widths, alpha=0.2)
nx.draw_networkx_edges(G, pos, edgelist=pos_edges, edge_color=pos_colors, width=pos_widths, alpha=0.8)




# --- Combined node drawing by shape and size ---
all_nodes = []
for node in G.nodes:
    if node in microbes:
        shape = 'o'  # 16S species
        size = 300 + (microbe_abund.get(node, 1) / max(microbe_abund)) * 1700
    elif node in microbe_ITS_nodes:
        shape = '^'  # ITS species
        size = 300 + (microbe_abund_ITS.get(node, 1) / max(microbe_abund_ITS)) * 1700

    elif node in metabolites:
        shape = 'D'  # metabolite
        size = (50 + (metabolite_abund.get(node, 1) / max(metabolite_abund)) * 1000)/1.5
    else:
        shape = 'o'
        size = 1
    all_nodes.append((node, size, shape))

# Sort and draw by shape and size (smallest last)
#all_nodes_sorted = sorted(all_nodes, key=lambda x: -x[1])

    
# for shape in ['o', '^', 'D']:
#     nodes_to_draw = [(n, s) for n, s, sh in all_nodes_sorted if sh == shape]
#     if not nodes_to_draw:
#         continue
#     nodelist = [n for n, _ in nodes_to_draw]
#     sizes = [s for _, s in nodes_to_draw]
#     nx.draw_networkx_nodes(
#         G, pos,
#         nodelist=nodelist,
#         node_color=[color_dict.get(n, 'gray') for n in nodelist],
#         node_size=sizes,
#         node_shape=shape,
#         edgecolors='black'
#     )
# Sort and draw all nodes globally by size, smallest last
all_nodes_sorted = sorted(all_nodes, key=lambda x: -x[1])  # smallest last

for node, size, shape in all_nodes_sorted:
    nx.draw_networkx_nodes(
        G, pos,
        nodelist=[node],
        node_color=[color_dict.get(node, 'gray')],
        node_size=[size],
        node_shape=shape,
        edgecolors='black'
    )



#nx.draw_networkx_labels(G, pos, font_size=9)

#plt.title("Tripartite CLR-weighted network", fontsize=14)
plt.title("Intra- and inter-kingdom co-correlation at 4 °C", fontsize=14)
plt.axis("off")

from matplotlib.lines import Line2D
from matplotlib.patches import Patch

# --- Shape legend (species vs metabolite) ---
shape_legend = [
    Line2D([0], [0], marker='o', color='w', label='Bacteria', markerfacecolor='black', markersize=10),
    Line2D([0], [0], marker='^', color='w', label='Fungi', markerfacecolor='black', markersize=10),
    Line2D([0], [0], marker='D', color='w', label='Metabolites', markerfacecolor='black', markersize=10)
]


# --- Edge legend ---
edge_legend = [
    Patch(color="#345084FF", alpha =0.8, label="Positive interaction"),
    Patch(color="#C0C0C0", alpha=0.2, label="Negative interaction")
]

# --- Color legend from color_dict ---
# Optional: provide nicer labels if needed
class_labels = {
    'Acetobacter cerevisiae': 'A. cerevisiae',
    'Lactobacillus sanfranciscensis': 'L. sanfranciscensis',
    'Pediococcus parvulus': 'P. parvulus',
    'Lactobacillus rossiae': 'L. rossiae',
    'Lactobacillus brevis': 'L. brevis',
    'Saccharomyces cerevisiae': 'S. cerevisiae',
    'Alternaria sp.': 'Alternaria sp.',
    'Aureobasidium pullulans': 'A. pullulans',
    'Pleosporaceae sp.': 'Pleosporaceae sp.',
    'Cladosporium herbarum': 'C. herbarum',
    'Alternaria rosae': 'A. rosae',
    'Pyrenophora tritici-repentis': 'P. tritici-repentis',
    'Pyrenophora variabilis': 'P. variabilis',
    'Stemphylium vesicarium': 'S. vesicarium',
    'Parastagonospora sp.': 'Parastagonospora sp.',
    'Amino acids, peptides, and analogues': 'Amino acids',
    'Carbohydrates and carbohydrate conjugates': 'Carbohydrates',
    'Fatty acyls': 'Fatty acyls',
    'Glycerolipids': 'Glycerolipids',
    'Glycerophospholipids': 'Glycerophospholipids',
    'Nucleosides, nucleotides, and analogues': 'Nucleotides',
    'Organic acids and derivatives': 'Organic acids',
    'Organoheterocyclic compounds': 'Organoheterocycles',
    'Other organic oxygen/nitrogen/sulfur compounds': 'Other organics',
    'Phenolic compounds': 'Phenolics',
    'Prenol lipids': 'Prenol lipids',
    'Steroids and steroid derivatives': 'Steroids'
}

# Build one Patch per unique color in the color_dict
color_legend = []
seen = set()
for label, color in color_dict.items():
    if color not in seen:
        seen.add(color)
        legend_label = class_labels.get(label, label)
        color_legend.append(Patch(facecolor=color, edgecolor=None, label=legend_label))

# --- Combine all legend elements ---
legend_elements = shape_legend  + edge_legend + color_legend

# --- Draw the legend ---
plt.legend(
    handles=legend_elements,
    loc='upper left',
    bbox_to_anchor=(1.02, 1),
    fontsize=12,
    borderaxespad=0.,
    title='',
    title_fontsize=10,
    ncol =2
)


plt.tight_layout()
# Save the figure
plt.savefig("networks/20250524_network_co_correlation_4C.pdf", bbox_inches='tight', dpi=300)
plt.show()
```
